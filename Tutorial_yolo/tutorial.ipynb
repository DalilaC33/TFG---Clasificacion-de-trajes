Predecir una imagen 

```python
import cv2
import torch
from ultralytics import YOLO

def predict_image(image_path, model_path=custom_model):
    # Cargar modelo
    model = YOLO(model_path)
    
    # Cargar imagen
    img = cv2.imread(image_path)
    if img is None:
        print("Error: No se pudo cargar la imagen.")
        return
    
    # Realizar predicción
    results = model(img, verbose=False)
    
    # Obtener la categoría con mayor probabilidad
    for r in results:
        conf_val = dict(enumerate(r.probs.data.numpy(), 1))
        names = r.names
        max_index = max(conf_val, key=conf_val.get)
        predicted_class = names[max_index]
        confidence = conf_val[max_index]
        
        print(f"Predicción: {predicted_class} con confianza {confidence:.2f}")
        return predicted_class, confidence

# Ejemplo de uso
image_path = "C:\\Users\\dalil\\OneDrive\\Imagens\\ins.png"
predict_image(image_path)
```

```python
import cv2
import numpy as np
import urllib.request

# Aquí debes poner la URL de tu imagen
url = "C:\\Users\\dalil\\Pictures\\Screenshots\\pop.png"

# Función para leer una imagen desde una URL
def url_to_image(url):
    with urllib.request.urlopen(url) as response:
        image = np.asarray(bytearray(response.read()), dtype="uint8")
        image = cv2.imdecode(image, cv2.IMREAD_COLOR)
        return image

img = url_to_image(url)

String = 'Results'
results = model(img, stream=True, verbose=False)
for r in results:
    conf_val = dict(enumerate(r.probs.data.numpy(), 1))
    names = r.names
    i = 1
    max_index = max(conf_val, key=conf_val.get)
    for (k1, v1), (k2, v2) in zip(names.items(), conf_val.items()):
        String = str(v1) + "->" + str(v2)
        if max_index == i:
            color = (0, 255, 0)
        else:
            color = (0, 0, 255)
        img = cv2.putText(img, String, (20, 35 * i), cv2.FONT_HERSHEY_SIMPLEX, 1.2, color, 2, cv2.LINE_AA)
        i += 1

cv2.imshow('Image from URL', img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
